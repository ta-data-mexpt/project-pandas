```python
# Step 2 Data Cleaning
# Eliminating missing values
data.isnull()
```

```python
columnas_nulas_suma = data.isnull().sum()
columnas_nulas_suma
```

```python
# Identify mean of missing values in percentage
columnas_nulas_mean = data.isnull().mean().round(4) * 100
columnas_nulas_mean
```

```python
# Drop columns where percentage of missing values is greater than 44%
data = data.drop(data.columns[data.isnull().mean()>0.44], axis=1)
```

```python
# Identify numerical columns where at least 90% of the values are the same
low_variance = []
 
for col in data._get_numeric_data():
    minimum = min(data[col])
    ninety_perc = np.percentile(data[col], 90)
    if ninety_perc == minimum:
        low_variance.append(col)
 
print(low_variance)
```

```python
# Data Manipulation
# Delete
data['Date'] = data['Date'].str.replace('[Reported, Letter dated]', '')
print(set(data['Date']))
```

```python
# Standardise month names if it is clear
data['Date'].replace(to_replace=[r'Jan?', r'Fe?b', r'Mar', r'Apr?', r'Ma?y', r'Ju?n', r'Ju?l', r'Au?g?', r'Se?p?', r'Oc?t?', r'No?v', r'Di?c', 'Bf'], 
           value=[r'Jan', r'Fb', r'Mar', r'Apr', r'My', r'Jn', r'Jul', r'Aug', r'S', r'Oc', r'Nv', r'Dc', 'Before '],
           regex=True, 
           inplace=True)
data['Date']
```

```python
# Rename columns
data = data.rename(columns={'pdf':'Pdf',
                            'href formula':'HREF Formula',
                            'href':'HREF',
                            'original order':'Original Order'})
data.columns
```

```python
# Change column order
column_names = ["Original Order", "Date", "Year", "Case Number", "Case Number.1", "Case Number.2", "Fatal(Y/N)", "Injury", "Type", "Surfing", "Country", "Area", "Location", "Name", "Sex", "Pdf", "HREF Formula", "HREF", "Investigator or Source"]
data = data.reindex(columns=column_names)
data
```

```python
# Sort by column "Original Order"
data.sort_values(by='Original Order', ascending = False)
```

```python
# Correct incorrect values by redistributing them within columns Fatal & Injury
fatal_injury=[x for x in data.Injury.unique() if re.findall('fatal',str(x).lower())]
fatal_index=[x for x in data[data.Injury.apply(lambda x:1 if x in fatal_injury else 0)==1].index]
for x in fatal_index:
        data.set_value(x,'Fatal (Y/N)','Y')
# Lamentablemente no me sale el codigo. De hecho hay tambien non-fatal.
```

```python
# Correct incorrect values by redistributing them within columns Sex and Name
male_name=[x for x in data.Name.unique() if re.findall('male',str(x).lower())]
male_index=[x for x in data[data.Name.apply(lambda x:1 if x in male_name else 0)==1].index]
for x in male_index:
    data.set_value(x,'Sex','M')
# Lamentablemente no me sale el codigo.
```

```python
#Step 4: Export Pandas DataFrame to a CSV file
data.to_csv("./Cleaned Data.csv", header=False)
```
