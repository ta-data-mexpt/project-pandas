```python
data=data.drop(columns=["Case Number","Year","Case Number.1", "Case Number.2","original order","Unnamed: 22","Unnamed: 23"])
```

```python
len(data.columns)
```

```python
row_index=[]
for i in range(0,len(data)):
    if data.iloc[i].count()<3:
        row_index.append(i)
```

```python
len(row_index)
```

```python
for i in row_index:
    data=data.drop(i)
```

```python
for i in range(0,len(data)):
    try:
        data.iloc[i,0]=pd.to_datetime(data.iloc[i,0])
    except:
        pass
```

```python
no_es_tiempo=[]
for i in range(0,len(data)):
    if not type(data.iloc[i,0]) == pd._libs.tslibs.timestamps.Timestamp:
        no_es_tiempo.append(i)
```

```python
len(no_es_tiempo)
```

```python
for i in no_es_tiempo:
    data.iloc[i,0]=re.sub("Reported(\s*)","",data.iloc[i,0])
    data.iloc[i,0]=re.sub("(S|s)ummer","06",data.iloc[i,0])
    data.iloc[i,0]=re.sub("(F|f)all","09",data.iloc[i,0])
    data.iloc[i,0]=re.sub("(W|w)inter","12",data.iloc[i,0])
    data.iloc[i,0]=re.sub("(S|s)pring","03",data.iloc[i,0])
    data.iloc[i,0]=re.sub("(E|e)arly","01",data.iloc[i,0])
    data.iloc[i,0]=re.sub("(L|l)ate","25",data.iloc[i,0])
    data.iloc[i,0]=re.sub("\.a","",data.iloc[i,0])
    data.iloc[i,0]=re.sub("\.b","",data.iloc[i,0])
    data.iloc[i,0]=re.sub("\.R","",data.iloc[i,0])
    data.iloc[i,0]=re.sub("(to have happened  \"on the weekend\")","",data.iloc[i,0])
    data.iloc[i,0]=re.sub("No date,(\s*)","",data.iloc[i,0])
    data.iloc[i,0]=re.sub("Before,(\s*)","",data.iloc[i,0])
    data.iloc[i,0]=re.sub("Before(\s*)","",data.iloc[i,0])
    data.iloc[i,0]=re.sub("Circa(\s*)","",data.iloc[i,0]) 
    data.iloc[i,0]=re.sub("Ca.(\s*)","",data.iloc[i,0]) 
    data.iloc[i,0]=re.sub("Letter dated","",data.iloc[i,0])  
    data.iloc[i,0]=re.sub("--","-",data.iloc[i,0])
    
```

```python
for i in range(0,len(data)):
    try:
        if not type(data.iloc[i,0]) == pd._libs.tslibs.timestamps.Timestamp:
            data.iloc[i,0]=pd.to_datetime(data.iloc[i,0])
    except:
        pass
```

```python
no_es_tiempo=[]
for i in range(0,len(data)):
    if not type(data.iloc[i,0]) == pd._libs.tslibs.timestamps.Timestamp:
        no_es_tiempo.append(i)
```

```python
def corrige_sexo(x):
    x=re.sub(".","",x)
    x=re.sub("nan","",x)
    x=re.sub("N","",x)
    x=re.sub("lli","F",x)
    x=re.sub("M ","M",x)
    return x
```
