```python
#Vemos que hay datos que tienen 'M ' que los trataremos como 'M' 
#y datos que tienen 'lli' que los trataremos como nulos = 0
```

```python
#Creamos la función para codificar los datos del sexo en el dataframe
def encode_sex (var):
    if var == 'M':
        return 0
    elif var == 'M ':
        return 0
    elif var == 'F':
        return 1
    elif var == 0:
        return 2
    elif var == 'lli':
        return 2
```

```python
#Aplicamos la función a nuestro data frame
data_2000s['Sex_encoded'] = data_2000s['Sex'].apply(encode_sex)

#Revisamos nuestro dataframe nuevamente y observamos que la columna de 'Sex_encoded' existe.
data_2000s
```

```python
#Si queremos probar que se cambió correctamente el filtro
#Debemos tener 1632 registros con el número 0. En el dataframe tenemos 1630 'M' y 2 'M '
#Debemos tener 341 registros con el número 1. En el dataframe tenemos 341 'F'
#Debemos de tener 105 registros con el número 2. En el dataframe tenemos 104 con valor '0' y 1 con valor 'lli'.

data_2000s['Sex_encoded'].value_counts()
```

```python
#Vamos a revisar los ataques de tiburones si son fatales o no

#Primero revisamos que valores tenemos en esta columna
data_2000s['Fatal (Y/N)'].value_counts()
```

```python
data_2000s['Fatal (Y/N)'].unique()
```

```python
#Creamos la función para codificar los datos de la fatalidad de los ataques en el dataframe
def encode_fatal (var):
    if var == 'N':
        return 0
    elif var == 'M':
        return 0
    elif var == 'Y':
        return 1
    elif var == 0:
        return 2
    elif var == 'UNKNOWN':
        return 2
    elif var == '2017':
        return 2
```

```python
#Aplicamos la función a nuestro data frame
data_2000s['Fatal_encoded'] = data_2000s['Fatal (Y/N)'].apply(encode_fatal)

#Revisamos nuestro dataframe nuevamente y observamos que la columna de 'Sex_encoded' existe.
data_2000s
```
